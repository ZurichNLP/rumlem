```python
# 1. Imports
from romansh_lemmatizer import Lemmatizer

# 2. Example text
text = """La mamma fa paun e la figlia legia in casa."""

# 3. Lemmatize
lemmatizer = Lemmatizer()
doc = lemmatizer(text)

# 4. Show idiom scores
print("Idiom probabilities:")
for idiom, score in sorted(doc.idiom_scores.items(), key=lambda x: x[1], reverse=True):
    print(f"{idiom.value:12s} {score:.3f}")

print("\nPredicted idiom:", doc.idiom.value)

# 5. Map each token → supported idioms
idioms = list(doc.idiom_scores.keys())

rows = []
for tok in doc.tokens:
    support = {idiom: "-" for idiom in idioms}
    for lemma in tok.all_lemmas.keys():
        support[lemma.idiom] = "✓"
    rows.append((tok.text, support))

# 6. Print table
header = ["Token"] + [idiom.value for idiom in idioms]
col_widths = [max(len(h), 8) for h in header]

# Header row
print("\n" + "  ".join(h.ljust(w) for h, w in zip(header, col_widths)))
print("-" * sum(col_widths))

# Token rows in sentence order
for token, support in rows:
    row = [token.ljust(col_widths[0])]
    for idiom, w in zip(idioms, col_widths[1:]):
        row.append(support[idiom].center(w))
    print("  ".join(row))
```
