## Classify Romansh Idiom
This script illustrates how we can use the romansh_lemmatizer to distinguish between the Romansh Idioms

### Preliminaries: Import

```python
# Imports
from romansh_lemmatizer import Lemmatizer
```

### Analyzing the Text
Initialise and apply the lemmatizer (non idiom-specific because we want to classify the text), then print the idiom scores (i.e., what percentage of words in the sentence did it find in the respective idiom).

```python

# Initialise non idiom-specific lemmatiser and lemmatize example text
text = """Uschia hai dà tant la dumonda da spustar la votaziun sco era la dumonda da votar en scrit."""

lemmatizer = Lemmatizer()
doc = lemmatizer(text)
```

```python
# Show idiom scores
print("Idiom scores:")
for idiom, score in sorted(doc.idiom_scores.items(), key=lambda x: x[1], reverse=True):
    print(f"{idiom.value:12s} {score:.3f}")

print("\nPredicted idiom:", doc.idiom.value)
```

### Pretty Print Results
Finally, print the results in a nice format.

```python
# Map each token → supported idioms
idioms = list(doc.idiom_scores.keys())

rows = []
for tok in doc.tokens:
    support = {idiom: "-" for idiom in idioms}
    for lemma in tok.all_lemmas.keys():
        support[lemma.idiom] = "✓"
    rows.append((tok.text, support))
```

```python
# Pretty print results
header = ["Token"] + [idiom.value for idiom in idioms]
col_widths = [max(len(h), 8) for h in header]

# Header row
print("\n" + "  ".join(h.ljust(w) for h, w in zip(header, col_widths)))
print("-" * sum(col_widths))

# Token rows in sentence order
for token, support in rows:
    row = [token.ljust(col_widths[0])]
    for idiom, w in zip(idioms, col_widths[1:]):
        row.append(support[idiom].center(w))
    print("  ".join(row))
```
