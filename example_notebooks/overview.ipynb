## Overview
This script serves as an overview to the workings of RUMLEM, with detailed descriptions on the object types and attributes.

### Importing the Lemmatizer

```python
from rumlem import Lemmatizer
```

### Initialising the Lemmatizer

```python
lemmatizer = Lemmatizer()
```

### Analyzing Text

```python
text = """La mamma fa paun e l'onda legia in casa."""
doc = lemmatizer(text)
```

This gives us an object of type "Doc". It has the following attributes:
- .text: the text we input
- .tokens: the tokens of that text
- .idiom: the idiom with the best coverage score
- .idiom_scores: a dictionary of Idiom objects and their score

```python
print(doc.tokens)
```

Each token is of type "Token":

```python
for tok in doc.tokens:
    print(tok, type(tok))
```

Tokens have three attributes:
- .text: the text content of the token
- .lemmas/.all_lemmas: These are dictionaries where the keys are of type "Lemma" and the values a list of "MorphAnalysis" objects. ".lemmas" gives all lemmas in the detected idiom, ".all_lemmas" gives all lemmas in any idiom.

```python
ind = 1
t = doc.tokens[ind]
print(f"Selected the first token for illustration purposes, '{t.text}'.")
print()
```

```python
ind = 1
t = doc.tokens[ind]
print(f"Selected the first token for illustration purposes, '{t.text}'.")
print()

print(".lemmas is of type:")
print(type(t.lemmas))  # <class 'dict'>
# Check first key/value pair to see what types it contains
first_key = next(iter(t.lemmas.keys()))
first_value = t.lemmas[first_key]
print(f"\tKey type: {type(first_key)}")      # e.g. <class '__main__.Lemma'>
print(f"\tValue type: {type(first_value)}")  # e.g. <class 'list'>
print(f"\t\tList element type: {type(first_value[0])}")  # <class '__main__.MorphAnalysis'>
```

Iteration over the lemmas of a token works as follows:

```python
print("Iterating over t.lemmas:") 
for lemma, morph_analyses in t.all_lemmas.items(): # the same for t.all_lemmas
    print(lemma, lemma.translation_de)
    for morph_analysis in morph_analyses:
        print("\t", morph_analysis) 
```

Note that objects of type "Lemma" have the attribute:
- .translation_de: this is the German translation of a lemma.
