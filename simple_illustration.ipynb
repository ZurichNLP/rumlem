```python
from romansh_lemmatizer import Lemmatizer
```

```python
text = """La mamma fa paun e la figlia legia in casa."""

lemmatizer = Lemmatizer()
doc = lemmatizer(text)
```

```python
print(doc.text)
print(doc.tokens)
print(doc.idiom)
print(doc.idiom_scores)
```

```python
for tok in doc.tokens:
    print(tok, type(tok))
```

```python
t = doc.tokens[1]
print(t.text)
print(type(t))
print()

print(type(t.lemmas))  # <class 'dict'>
# Check first key/value pair to see what types it contains
first_key = next(iter(t.lemmas.keys()))
first_value = t.lemmas[first_key]
print(f"Key type: {type(first_key)}")      # e.g. <class '__main__.Lemma'>
print(f"Value type: {type(first_value)}")  # e.g. <class 'list'>
print(f"List element type: {type(first_value[0])}")  # <class '__main__.MorphAnalysis'>
print()


for lemma, morph_analyses in t.lemmas.items():
    print(lemma, lemma.translation_de)
    for morph_analysis in morph_analyses:
        print("\t", morph_analysis)  # This calls MorphAnalysis.__str__, showing features like "Feat1=val1;Feat2=val2"
```
